```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

cocoa_beans = '/content/Global import of cocoa beans.xlsx'
cocoa_df = pd.read_excel(cocoa_beans)
cocoa_df.head()
```

```python
country_filter = cocoa_df['Region'] == 'Asia & Oceania'
filtered_cocoa = cocoa_df.loc[country_filter, :]
filtered_cocoa
```

```python
year = cocoa_df.groupby(['Year'])['Import in tonnes'].nunique()
numYears = np.array(year.index)
numImports = np.array(year.values)

plt.style.use('seaborn')
plt.plot(numYears, numImports, c= 'k')
plt.xlabel('Year', fontsize = 16)
plt.ylabel('Import in tonnes', fontsize = 16)
plt.title('Number of Asian Cocoa Imports', fontsize = 16)
```
